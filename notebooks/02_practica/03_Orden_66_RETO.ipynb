## 4. Preparar los datos

Usaremos el dataset **Digits** de Scikit-learn:

- contiene 1.797 imágenes de números escritos a mano;
- cada imagen mide 8 × 8 píxeles;
- cada fila se transforma en 64 características numéricas;
- el objetivo es clasificar cada imagen como un dígito del 0 al 9.

Este problema es visual y permite interpretar mejor la matriz de confusión y el informe de clasificación.

````python
data = load_digits(as_frame=True)

X = data.data
y = data.target

test_size = 0.20
random_state = 42

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=test_size,
    random_state=random_state,
    stratify=y,
)

print("Dataset: Digits")
print("Muestras totales:", len(X))
print("Entrenamiento:", X_train.shape)
print("Test:", X_test.shape)
print("Clases:", list(data.target_names))

# Mostramos una imagen para entender el problema.
plt.imshow(data.images[0], cmap="gray")
plt.title(f"Dígito real: {data.target[0]}")
plt.axis("off")
plt.show()
````
